IMPORT KISIMLARI

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
import matplotlib.pyplot as plt
import time
from sklearn import metrics
import seaborn
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import time
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
```

DOSYA AKTARMA

```python
dosya_isim="fake_news_dataset.csv"
data = pd.read_csv('fake_news_dataset.csv')
```

TF-IDF

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=4000)
X_text = vectorizer.fit_transform(data['tweet'])
y_classification = data['target']
```

SINIFLANDIRMALAR

```python
knn3e = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn5e = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn3m = KNeighborsClassifier(n_neighbors=3, metric='minkowski')
knn5m = KNeighborsClassifier(n_neighbors=5, metric='minkowski')
DTEntropy = DecisionTreeClassifier(criterion='entropy')
DTGini = DecisionTreeClassifier(criterion='gini')
NB = GaussianNB()
SVM = SVC(kernel='linear')
```

train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X_text, y_classification, test_size=0.2, random_state=42)
```

knn3e için accuracy, f_measure, precision, recall, training time, testing time ve karmaşıklık matrisi(confusion matrix)

```python
t = time.time()
knn3e.fit(X_train, y_train)
print("training time:",time.time()-t)
t = time.time()
knn3e.score(X_test,y_test)
print("Testing time:",time.time()-t)
y_pred = knn3e.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f_measure = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
print("accuracy:",accuracy)
print("f_measure:", f_measure)
print("precision:",precision)
print("recall:",recall)
confusion_matrix = metrics.confusion_matrix(y_test,y_pred)
seaborn.heatmap(confusion_matrix,cmap="Greys",annot=True)
plt.show()
```
